```python
import numpy as np
```

```python
a = np.array([[0, 1/2, 1, 0],
            [1/3, 0, 0, 1/2],
            [1/3, 0, 0, 1/2],
            [1/3, 1/2, 0, 0]])
```

```python
a_leak = np.array([[0, 0, 0, 1/2],
                [0, 0, 0, 1/2],
                [0, 1, 0, 0],
                [0, 0, 1, 0]])
```

```python
a_sink = np.array([[0, 0, 0, 0],
                [1/2, 0, 0, 1],
                [0, 1, 1, 0],
                [1/2, 0, 0, 0]])
```

```python
b = np.array([1/4, 1/4, 1/4, 1/4])
w = b
```

```python
def work(a, w):
    for i in range(100):
        w = np.dot(a, w)
        print(w)
```

```python
def random_work(a, w, n):
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d*np.dot(a, w)
        print(w)
```

```python
work(a, w)
```

```python
random_work(a, w, 4)
```

```python
random_work(a_leak, w, 4)
```

```python
random_work(a_sink, w, 4)
```
